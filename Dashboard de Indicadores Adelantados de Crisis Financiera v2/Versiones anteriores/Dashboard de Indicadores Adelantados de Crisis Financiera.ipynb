Manual de Usuario e Interpretación Profesional (Versión Corregida)
1. Spread de Bonos de Alto Rendimiento (High-Yield)
¿Qué es? Es el "salario extra" que exigen los inversores por prestarle dinero a las empresas más riesgosas en lugar de al gobierno de EE.UU. (que se considera libre de riesgo).

Parámetro Normal (🟢 < 4%): Cuando el spread es bajo, significa que los inversores están confiados. Piensan: "La economía va bien, estas empresas no van a quebrar, así que no necesito que me paguen mucho más por el riesgo". Hay "apetito por el riesgo".

Parámetro Anómalo (🔴 > 6%): Un spread alto es una señal de miedo. Los inversores piensan: "La economía se debilita, el riesgo de que estas empresas no me paguen es alto, así que o me pagas una prima de riesgo enorme, o no te presto mi dinero".

Cómo Validarlo: La señal es más fuerte si la subida es rápida y vertical. Un aumento lento puede ser un simple ajuste del ciclo económico. Una subida explosiva, como la que se vio en Septiembre de 2008 o Marzo de 2020, es una sirena de bomberos. Valídala con el VIX; si ambos se disparan a la vez, el mercado está entrando en pánico.

2. La Curva de Tipos (Yield Curve)
¿Qué es? Es la diferencia entre lo que paga un bono del gobierno a 10 años y lo que paga uno a 2 años.

Parámetro Normal (🟢 > 0.25%): Normalmente, prestar a largo plazo es más riesgoso (pueden pasar más cosas en 10 años que en 2), así que debería pagar más. Esto se llama una "curva con pendiente positiva" y es señal de una economía sana que espera crecer.

Parámetro Anómalo (🔴 < 0%): Se llama "inversión de la curva". Ocurre cuando el bono a 2 años paga MÁS que el de 10 años. Es una señal perversa. El mercado está gritando que el futuro a corto plazo (próximos 2 años) es tan incierto y riesgoso que prefiere "amarrar" su dinero por 10 años aunque le paguen menos. Es el predictor de recesiones más fiable que existe, aunque puede tardar hasta 2 años en materializarse.

Cómo Validarlo: Una inversión de un solo día puede ser ruido. La señal se valida por su persistencia y profundidad. Si la curva se mantiene invertida durante varias semanas o meses, y la inversión supera los -0.25% o -0.50%, la señal es extremadamente robusta. No es una señal para vender todo mañana, sino para empezar a construir una estrategia defensiva para los próximos 12-18 meses.

3. El Índice VIX ("Índice del Miedo")
¿Qué es? No mide la volatilidad de hoy, sino la que el mercado espera para los próximos 30 días. Se calcula a partir de los precios de las opciones del S&P 500. Es, literalmente, el precio del "seguro" contra una caída del mercado.

Parámetro Normal (🟢 < 20): Cuando el seguro es barato, nadie tiene miedo. Es un entorno de calma y complacencia, a menudo asociado con mercados alcistas estables.

Parámetro Anómalo (🔴 > 30): Cuando el seguro se vuelve carísimo, es porque todo el mundo quiere comprarlo a la vez. Esto solo ocurre cuando hay pánico y los inversores temen una caída inminente.

Cómo Validarlo: La señal de peligro tiene dos formas principales. Primero, un pico súbito, que es lo que el programa detecta (un aumento de más del 50% sobre su media reciente). Esto indica un shock o una noticia que ha asustado al mercado repentinamente. Segundo, un VIX que se mantiene elevado por encima de 25-30 durante varios días o semanas; esta señal es aún más potente, ya que indica que el miedo no es pasajero, sino que se está instalando en el mercado.

4. El Índice del Dólar (DXY)
¿Qué es? Mide la fuerza del dólar estadounidense contra una canasta de las principales monedas del mundo (Euro, Yen, etc.).

Parámetro Normal (🟢 Cambio < 3% en 3 meses): Movimientos suaves y tendenciales.

Parámetro Anómalo (🔴 Cambio > 6% en 3 meses): Una subida vertical del dólar. En una crisis global, el mundo entero entra en pánico y solo quiere una cosa: el activo considerado más seguro del planeta, el dólar estadounidense. Venden acciones, bonos, propiedades en sus países y compran dólares. Esto se conoce como la "Huida hacia la Calidad" o flight to quality.

Cómo Validarlo: Una subida del DXY por sí sola no es una señal de crisis; podría deberse a que la Reserva Federal está subiendo los tipos más rápido que otros bancos centrales. La señal inequívoca de crisis es cuando la subida del DXY ocurre simultáneamente con una caída de las bolsas mundiales, una ampliación de los spreads de crédito (Indicador 1) y una caída de los precios de las materias primas. Es la confluencia lo que valida la señal de pánico global.

```python
import pandas as pd
import yfinance as yf
from fredapi import Fred
from datetime import datetime, timedelta
import os
import numpy as np  # Requerido por la función del usuario
import logging  # Requerido por la función del usuario
import warnings  # Requerido por la función del usuario

# -----------------------------------------------------------------------------
# CONFIGURACIÓN INICIAL Y DE SEGURIDAD
# -----------------------------------------------------------------------------

# (Configuración de logging básica para que funcione la función de descarga)
logging.basicConfig(level=logging.INFO, format='%(levelname)s: %(message)s')

# (Supresores de warnings provistos por el usuario)
warnings.filterwarnings("ignore", category=FutureWarning, module='yfinance.download')
warnings.filterwarnings("ignore", category=FutureWarning, module='yfinance.utils')

# (Mantenemos el método "inseguro" solicitado por el usuario)
FRED_API_KEY = os.environ.get('FRED_API_KEY', '6c4e4e9a1b4589bc2a0639ce3a5e0fba')
fred = Fred(api_key=FRED_API_KEY)

# Fechas para el análisis histórico
END_DATE = datetime.now()
CRISIS_START_DATE = '1970-01-01'

# Definición de períodos de crisis históricas para el Stress Test
CRISIS_PERIODS = {
    'Crisis del Tequila (México)': ('1994-12-01', '1995-05-01'),
    'Crisis Financiera Asiática': ('1997-07-01', '1998-05-01'),
    'Crisis del Rublo (Rusia)': ('1998-08-01', '1999-01-01'),
    'Burbuja .com (Explosión)': ('2000-03-01', '2001-09-01'),
    'Crisis Financiera Global (GFC)': ('2008-09-01', '2009-03-01'),
    'Crisis de Deuda Europea': ('2011-08-01', '2012-06-01'),
    'Crash por COVID-19': ('2020-02-15', '2020-04-01'),
}

# -----------------------------------------------------------------------------
# FUNCIÓN DE DESCARGA (PROVISTA POR EL USUARIO)
# -----------------------------------------------------------------------------

def download_single_ticker_data(ticker, start_date, end_date):
    """
    Función de descarga robusta de yfinance (vía usuario)
    Maneja la nueva lógica de yfinance y auto_adjust=True.
    """
    try:
        data = yf.download(ticker, start=start_date, end=end_date, auto_adjust=True, progress=False)
        if data.empty:
            logging.warning(f"No hay datos para {ticker} en el rango especificado. Intentando descargar sin rango.")
            data = yf.download(ticker, auto_adjust=True, progress=False)

        if data.empty:
             logging.warning(f"No hay datos para {ticker} ni siquiera sin rango.")
             return pd.Series(dtype=float)

        # yfinance 0.2.x+ descarga en un DataFrame con MultiIndex (si son >1 tickers)
        # o un DataFrame simple (si es 1 ticker).
        if isinstance(data.columns, pd.MultiIndex):
            if 'Close' in data.columns:
                prices_series = data['Close'].copy()
            elif 'Adj Close' in data.columns: # auto_adjust=True no suele tener 'Adj Close', pero es buena cobertura
                prices_series = data['Adj Close'].copy()
            else:
                logging.warning(f"Ni 'Adj Close' ni 'Close' encontrados (MultiIndex) para {ticker}.")
                return pd.Series(dtype=float)
        elif 'Close' in data.columns: # Caso de un solo ticker sin MultiIndex
            prices_series = data['Close'].copy()
        elif 'Adj Close' in data.columns:
            prices_series = data['Adj Close'].copy()
        else:
            logging.warning(f"Ni 'Adj Close' ni 'Close' encontrados (SingleIndex) para {ticker}.")
            return pd.Series(dtype=float)

        return prices_series.dropna().astype(np.float64)
    except Exception as e:
        logging.error(f"Error al descargar {ticker}: {str(e)}")
        return pd.Series(dtype=float) # Devuelve Series vacía en lugar de None

# -----------------------------------------------------------------------------
# FUNCIONES AUXILIARES (ROBUSTEZ v2)
# -----------------------------------------------------------------------------

def sanitize_series(series, series_name):
    """ Limpia y prepara las series temporales. """
    series.name = series_name
    series_clean = series.asfreq('B').ffill().dropna()
    return series_clean

def get_historical_stress_value(series, crisis_name, indicator_type='max'):
    """ Obtiene el valor máximo/mínimo durante una crisis, devolviendo 'N/A' si no hay datos. """
    start, end = CRISIS_PERIODS[crisis_name]
    crisis_data = series.loc[start:end]
    if crisis_data.empty:
        return "N/A"

    if indicator_type == 'min':
        val = crisis_data.min()
    else:
        val = crisis_data.max()

    if pd.isna(val):
        return "N/A"
    return val

# -----------------------------------------------------------------------------
# FUNCIONES DE ANÁLISIS POR INDICADOR (CORREGIDAS v5)
# -----------------------------------------------------------------------------

def safe_format(value, suffix=""):
    """ Formatea un valor como .2f si es número, o lo devuelve como string ("N/A") si no lo es. """
    if isinstance(value, (int, float)):
        return f"{value:.2f}{suffix}"
    return str(value)

def check_high_yield_spread():
    """ Indicador 1: Prima de Riesgo de Bonos Basura """
    try:
        series_raw = fred.get_series('BAMLH0A0HYM2', observation_start=CRISIS_START_DATE, end_date=END_DATE)
        series = sanitize_series(series_raw, 'BAMLH0A0HYM2')

        if series.empty:
            return {"level": -1, "explanation": "Error: Datos de High Yield Spread vacíos."}

        latest_value = series.iloc[-1]
        val_3m_ago = series.asof(END_DATE - pd.DateOffset(months=3))

        status = ""
        level = 0

        if latest_value < 4:
            status = "🟢 NORMAL: Apetito por el riesgo saludable."
            level = 0
        elif 4 <= latest_value < 6:
            status = "🟡 ALERTA: Aumento de la aversión al riesgo."
            level = 1
        else:
            status = f"🔴 PELIGRO: Estrés severo en el mercado de crédito."
            level = 2

        if val_3m_ago != 0 and latest_value > val_3m_ago * 1.25:
            status += "\n   VALIDACIÓN: El aumento ha sido RÁPIDO (>25% en 3 meses), lo que refuerza la señal."

        gfc_peak = get_historical_stress_value(series, 'Crisis Financiera Global (GFC)', 'max')
        covid_peak = get_historical_stress_value(series, 'Crash por COVID-19', 'max')

        explanation = f"""
        [ Spread Bonos Alto Rendimiento: {latest_value:.2f}% ]
        {status}
        - STRESS TEST: Pico en Crisis 2008: {safe_format(gfc_peak, '%')} | Pico en Crash COVID-19: {safe_format(covid_peak, '%')}
        """
        return {"level": level, "explanation": explanation}

    except Exception as e:
        return {"level": -1, "explanation": f"Error al obtener datos del High Yield Spread: {e}"}

def check_yield_curve():
    """ Indicador 2: Curva de Tipos """
    try:
        series_raw = fred.get_series('T10Y2Y', observation_start=CRISIS_START_DATE, end_date=END_DATE)
        series = sanitize_series(series_raw, 'T10Y2Y')

        if series.empty:
            return {"level": -1, "explanation": "Error: Datos de Curva de Tipos vacíos."}

        latest_value = series.iloc[-1]

        status = ""
        level = 0

        if latest_value > 0.25:
            status = f"🟢 NORMAL: Curva con pendiente positiva saludable."
            level = 0
        elif 0 <= latest_value <= 0.25:
            status = f"🟡 ALERTA: Aplanamiento de la curva."
            level = 1
        else:
            status = f"🔴 PELIGRO: INVERSIÓN DE LA CURVA. Fuerte predictor de recesión."
            level = 2

        window_data_3m = series.loc[END_DATE - pd.DateOffset(months=3) : END_DATE]
        if not window_data_3m.empty:
            inversion_pct = (window_data_3m < 0).mean()
            if inversion_pct >= 0.80:
                status += f"\n   VALIDACIÓN: La inversión es PERSISTENTE ({inversion_pct:.0%} de los últimos 3 meses), aumenta la señal."

        dotcom_min = get_historical_stress_value(series, 'Burbuja .com (Explosión)', 'min')
        gfc_min = get_historical_stress_value(series, 'Crisis Financiera Global (GFC)', 'min')
        covid_min = get_historical_stress_value(series, 'Crash por COVID-19', 'min')

        explanation = f"""
        [ Curva de Tipos (10Y-2Y): {latest_value:.2f}% ]
        {status}
        - STRESS TEST: Mínimo pre-.com: {safe_format(dotcom_min, '%')} | Mínimo pre-GFC 2008: {safe_format(gfc_min, '%')} | Mínimo pre-COVID: {safe_format(covid_min, '%')}
        """
        return {"level": level, "explanation": explanation}

    except Exception as e:
        return {"level": -1, "explanation": f"Error al obtener datos de la Curva de Tipos: {e}"}

def check_vix():
    """ Indicador 3: Índice de Volatilidad VIX """
    try:
        # (SOLUCIÓN v5: Se utiliza la función de descarga del usuario)
        series_raw = download_single_ticker_data('^VIX', CRISIS_START_DATE, END_DATE.strftime('%Y-%m-%d'))

        if series_raw.empty:
            return {"level": -1, "explanation": "Error: Datos del VIX vacíos (función de descarga falló)."}

        series = sanitize_series(series_raw, 'VIX')

        if series.empty:
            return {"level": -1, "explanation": "Error: Datos del VIX vacíos (post-saneamiento)."}

        latest_value = series.iloc[-1]

        status = ""
        level = 0

        if latest_value < 20:
            status = "🟢 NORMAL: Complacencia en el mercado."
            level = 0
        elif 20 <= latest_value < 30:
            status = "🟡 ALERTA: Miedo creciente."
            level = 1
        else:
            status = "🔴 PELIGRO: Pánico en el mercado."
            level = 2

        window_data_1m = series.loc[END_DATE - pd.DateOffset(months=1) : END_DATE]
        if not window_data_1m.empty:
            avg_1m = window_data_1m.mean()
            # Fix: Ensure avg_1m is a scalar before comparison
            # Explicitly convert to float to avoid Series comparison issue
            avg_1m_scalar = float(avg_1m) if isinstance(avg_1m, (int, float)) or (isinstance(avg_1m, pd.Series) and not avg_1m.empty) else np.nan
            if not pd.isna(avg_1m_scalar) and avg_1m_scalar != 0 and latest_value > avg_1m_scalar * 1.5:
                status += "\n   VALIDACIÓN: El aumento del miedo ha sido SÚBITO (>50% sobre la media del mes)."


        gfc_peak = get_historical_stress_value(series, 'Crisis Financiera Global (GFC)', 'max')
        covid_peak = get_historical_stress_value(series, 'Crash por COVID-19', 'max')

        explanation = f"""
        [ Índice VIX: {latest_value:.2f} ]
        {status}
        - STRESS TEST: Pico en Crisis 2008: {safe_format(gfc_peak)} | Pico en Crash COVID-19: {safe_format(covid_peak)}
        """
        return {"level": level, "explanation": explanation}

    except Exception as e:
        return {"level": -1, "explanation": f"Error al obtener datos del VIX: {e}"}

def check_dollar_index():
    """ Indicador 4: Índice del Dólar (DXY) """
    series_raw = None
    ticker_used = ""
    try:
        # Intento 1: FRED
        series_raw = fred.get_series('DTWEXM', observation_start=CRISIS_START_DATE, end_date=END_DATE)
        ticker_used = "Dólar (FRED: DTWEXM)"
    except Exception as e_fred:
        logging.warning(f"Advertencia: Falló al obtener DTWEXM de FRED ({e_fred}). Intentando fallback con Yahoo Finance...")
        try:
            # Intento 2: Yahoo Finance (Fallback)
            # (SOLUCIÓN v5: Se utiliza la función de descarga del usuario)
            series_raw = download_single_ticker_data('DX=F', CRISIS_START_DATE, END_DATE.strftime('%Y-%m-%d'))

            if series_raw.empty:
                 return {"level": -1, "explanation": "Error: Datos del DXY (Yahoo) vacíos."}

            ticker_used = "Dólar (Yahoo: DX=F)"
        except Exception as e_yf:
            return {"level": -1, "explanation": f"Error al obtener datos del DXY de FRED y Yahoo: {e_yf}"}

    series = sanitize_series(series_raw, ticker_used)

    if series.empty:
            return {"level": -1, "explanation": "Error: Datos del Dólar vacíos."}

    latest_value = series.iloc[-1]
    val_3m_ago = series.asof(END_DATE - pd.DateOffset(months=3))

    if val_3m_ago == 0 or pd.isna(val_3m_ago):
        return {"level": -1, "explanation": "Error: No hay datos del Dólar de hace 3 meses para comparar."}

    change_3m = (latest_value / val_3m_ago) - 1

    status = ""
    level = 0

    if change_3m < 0.03:
        status = "🟢 NORMAL: Movimientos del dólar estables."
        level = 0
    elif 0.03 <= change_3m < 0.06:
        status = "🟡 ALERTA: Fortalecimiento notable del dólar (Huida hacia la calidad moderada)."
        level = 1
    else:
        status = "🔴 PELIGRO: Fortalecimiento SÚBITO y VERTICAL del dólar. Pánico global."
        level = 2

    status += f"\n   Análisis: El Dólar se apreció un {change_3m:.2%} en los últimos 3 meses."

    gfc_peak = get_historical_stress_value(series, 'Crisis Financiera Global (GFC)', 'max')
    covid_peak = get_historical_stress_value(series, 'Crash por COVID-19', 'max')

    explanation = f"""
    [ {ticker_used}: {latest_value:.2f} ]
    {status}
    - STRESS TEST: Pico en Crisis 2008: {safe_format(gfc_peak)} | Pico en Crash COVID-19: {safe_format(covid_peak)}
    """
    return {"level": level, "explanation": explanation}

# -----------------------------------------------------------------------------
# EJECUCIÓN DEL DASHBOARD
# -----------------------------------------------------------------------------

def run_dashboard():
    """
    Ejecuta todos los chequeos y presenta un resumen del dashboard.
    """
    print("==========================================================")
    print(f"=   DASHBOARD DE INDICADORES ADELANTADOS DE CRISIS (v5)  =")
    print(f"=   Fecha del Análisis: {END_DATE.strftime('%Y-%m-%d')}                =")
    print("==========================================================")

    indicators = [
        check_high_yield_spread,
        check_yield_curve,
        check_vix,
        check_dollar_index
    ]

    total_level = 0
    results = []

    for check_func in indicators:
        result = check_func()
        results.append(result)
        if result['level'] > 0:
            total_level += result['level']
        print(result['explanation'])
        print("-" * 58)

    print("\n==========================================================")
    print(f"=   RESUMEN Y NIVEL DE ALERTA GENERAL                  =")
    print("==========================================================")

    if total_level == 0:
        print("🟢 NIVEL DE ALERTA GENERAL: BAJO")
        print("Todos los indicadores se encuentran en rangos de normalidad. El entorno macroeconómico parece estable.")
    elif total_level <= 3:
        print("🟡 NIVEL DE ALERTA GENERAL: MEDIO")
        print("Algunos indicadores clave muestran señales de estrés. Se recomienda cautela, aumentar la vigilancia y reducir la exposición al riesgo.")
    else: # total_level > 3
        print("🔴 NIVEL DE ALERTA GENERAL: ALTO")
        print("Múltiples indicadores críticos están en zona de peligro. La probabilidad de un evento de crisis financiera o recesión es elevada.")

    print("==========================================================")

# -----------------------------------------------------------------------------
# EJECUCIÓN DEL PROGRAMA
# -----------------------------------------------------------------------------
if __name__ == '__main__':
    if FRED_API_KEY == 'TU_CLAVE_API_AQUI_SI_NO_ESTA_EN_ENTORNO':
        print("Error: No se encontró la clave de API de FRED.")
    else:
        run_dashboard()
```
